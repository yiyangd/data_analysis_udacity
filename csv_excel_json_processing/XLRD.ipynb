```python
import xlrd
datafile = "2013_ERCOT_Hourly_Load_Data.xls"

def parse_file(datafile):
    workbook = xlrd.open_workbook(datafile) # open the workbook
    sheet = workbook.sheet_by_index(0) # 0-based indexing

    data = [[sheet.cell_value(r, col) 
                for col in range(sheet.ncols)] 
                    for r in range(sheet.nrows)]
    
    # column slicing trick, only sheet has column value
    # data.col_value was wrong because data is list
    cv = sheet.col_values(1,start_rowx=1,end_rowx=None)
    maxval = max(cv)
    minval = min(cv)
    maxpos = cv.index(maxval) + 1
    minpos = cv.index(minval) + 1
    # because index starts by 0 so maxpos and minpos must add 1 to find time
    
    maxtime = sheet.cell_value(maxpos,0)
    realmaxtime = xlrd.xldate_as_tuple(maxtime,0)
    mintime = sheet.cell_value(minpos,0)
    realmintime = xlrd.xldate_as_tuple(mintime,0)
    
    data = {
            'maxtime' : realmaxtime,
            'maxvalue': maxval,
            'mintime' : realmintime,
            'minvalue': minval,
            'avgcoast': sum(cv)/float(len(cv))
    }
    
    return data
data = parse_file(datafile)




import pprint
pprint.pprint(data)
```
