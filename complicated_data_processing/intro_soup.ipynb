```python
from bs4 import BeautifulSoup
```

```python
def options(soup,id):
    option_values = []
    carrier_list = soup.find(id = id)
    print("carrier_list:\n",carrier_list)
    for option in carrier_list.find_all('option'):  #find option tag under carrierlist and pullout the value
        option_values.append(option['value'])
    return option_values

def print_list(label, codes):
    print ("\n%s:" % label)
    for c in codes:
        print(c)
        
def main():
    soup = BeautifulSoup(open("virgin_and_logan_airport.html"))
    
    codes = options(soup, 'CarrierList') # id = CarrierList
    print(codes)
    print_list("Carriers",codes)
    codes = options(soup,'AirportList')
    print_list("Airports", codes)
```

```python
main()
```
